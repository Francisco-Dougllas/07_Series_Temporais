# Criação de uma Série Temporal Aleatoria com Distribuição Normal

```python
# Importando as Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
from matplotlib.pylab import rcParams # Atribuo uma configuração a todos os graficos
rcParams['figure.figsize'] = 15, 6
```

## Série Anual

```python
# Serie de 1980 a 2022
# (0, 1, 43) 0 Media, 1, Desvio padrao, 43 sao os valores que desejo
np.random.seed(10)
dados1 = np.random.normal(0, 1, 43)
dados1
```

```python
# Tipo
type(dados1)
```

```python
# Serie
serie = pd.Series(dados1)
serie[:10]
```

```python
# Tipo de serie
type(serie)
```

```python
# Visualizando grafico
serie.plot();
```

```python
# COnvertendo para dataframe
dados1 = pd.DataFrame(dados1)
dados1[:10]
```

```python
# Renomeando a Coluna
dados1.columns = ['Valores']
dados1.head()
```

```python
# Shape
dados1.shape
```

```python
# Descricao estatistica
dados1.describe()
```

```python
# Criando o indice Data
indice = pd.date_range('1980', periods=len(dados1), freq='Y')
indice
```

```python
# Criando uma nova serie
serie1 = pd.Series(dados1['Valores'].values, index=indice )
serie1[:10]
```

```python
# Tipo de serie1
type(serie1)
```

```python
# Plotando um novo grafico
serie1.plot();
```

```python
# Import da Biblioteca scipy estatistica
import scipy.stats as stats

# Com essa biblioteca de estatistica podemos saber se a distribuição e ou não normal
```

```python
# Distribuição normal dos Dados em forma grafica
stats.probplot(serie1, dist='norm', plot=plt)
plt.title('Normal QQ Plot');
```

**Nesse Grafico temos uma distribuição normal dos dados**

## Teste Shapiro-Wilk
    Criterios:
        Nivel de Significância de 0,05 ou 5% (Mais Utilizado)
        Quando p > 0,05 (Distribuição normal)

```python
# Testando
e, p = stats.shapiro(serie1)
print(f'Estatistica de teste: {e}')
print(f'P-Valor: {p}')
```
