# CRIAÇÃO DE UM PASSEIO ALEATORIO

```python
# Import das Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
# Atribuindo a Imagem grafica
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,6
```

```python
# Criando os dados aleatorios
from random import sample, random
dados1 = sample(range(100), k=41)
print(dados1, end=' ')
```

```python
# Criando a variaval ano
anos = list(range(1980, 2021))
print(anos, end=' ')
```

```python
# Criando a serie
serie1 = pd.Series(dados1, index=anos)
serie1
```

```python
# Plot serie
serie1.plot();
```

## Serie Passeio Aleatorio

```python
serie2 = list()

serie2.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movimento = -1 if random() < 0.5 else 1
    valor = serie2[i-1] + movimento
    serie2.append(valor)
```

```python
# Criando a serie e plotando o grafico
serie2 = pd.Series(serie2)
serie2.plot();
```
