# Media Movel

```python
# Import das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 5
```

**Desvio da temperatura media global da terra-oceano (com frequencia a media de 1951-1980), medidos em graus Celsius, para os anos de 1880-2015**

```python
# Importaçao dos dados
temperatura = pd.read_csv("temp_global.csv")
temperatura
```

```python
# Criando o indice
indice = pd.date_range('1880 Jan 1', periods = len(temperatura), freq='Y')
indice
```

```python
# Criando a serie
serie = pd.Series(temperatura['x'].values, index=indice)
serie
```

```python
# Plot grafico
serie.plot();
```

```python
# Criando a media movel
media_movel1 = serie.rolling(window=3)
media_movel1 = media_movel1.mean()
```

```python
# Plot
plt.plot(media_movel1);
```

```python
# Plotando o grafico completo
plt.plot(serie, label='Série Real')
plt.plot(media_movel1, color='red', label='Media Movel 3 Anos')
plt.legend(loc='best');
```

```python
# Criando a Media movel 2
media_movel2 = serie.rolling(window=10)
media_movel2 = media_movel2.mean()
```

```python
# Plot completo
plt.plot(serie, label='Série Real')
plt.plot(media_movel2, color='red', label='Media Movel 10 Anos')
plt.legend(loc='best');
```

```python
# Criando a Media movel 3 com janelas de 20 
media_movel3 = serie.rolling(window=20)
media_movel3 = media_movel3.mean()
```

```python
# Plot completo
plt.plot(serie, label='Série Real')
plt.plot(media_movel3, color='red', label='Media Movel 20 Anos')
plt.legend(loc='best');
```

```python
# Grafico com todas as medias moveis
plt.plot(serie, label='Série Real')
plt.plot(media_movel1, color='orange', label='Media Movel 3 Anos')
plt.plot(media_movel2, color='red', label='Media Movel 10 Anos')
plt.plot(media_movel3, color='green', label='Media Movel 20 Anos')
plt.legend(loc='best');
```

```python
# Criando uma media movel centralizada
media_movel4 = serie.rolling(window=3, center=True)
media_movel4 = media_movel4.mean()
```

```python
media_movel5 = serie.rolling(window=10, center=True)
media_movel5 = media_movel5.mean()
```

```python
media_movel6 = serie.rolling(window=20, center=True)
media_movel6 = media_movel5.mean()
```

```python
# Grafico com todas as medias moveis
plt.plot(serie, label='Série Real')
plt.plot(media_movel4, color='orange', label='Media Movel 3 Anos')
plt.plot(media_movel5, color='red', label='Media Movel 10 Anos')
plt.plot(media_movel6, color='green', label='Media Movel 20 Anos')
plt.legend(loc='best');
```
